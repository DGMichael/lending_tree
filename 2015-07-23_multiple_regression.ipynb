```python
#TRAIN MODEL WITH ANNUAL INCOME AS INPUT VARIABLE:
est = smf.ols(formula='Interest ~ Annual', data=df).fit()
est.summary()
```

```python
#Train Model with two input variables:
df["home_own"] = df["Home.Ownership"]
est2 = smf.ols(formula='Interest ~ Annual + C(home_own)', data=df).fit()
est2.summary()
```

```python
#Allow interaction between home ownership variable and annual income
est3 = smf.ols(formula='Interest ~ Annual * C(home_own)', data=df).fit()
est3.summary()
```

```python
#Attempt multiple regression using ols(y,X) syntax:
y = df["Interest"]
X1 = pd.Categorical(df["home_own"]).labels
X2 = df["Annual"]
#X1 should be a binary array of k-1 size...but it is not.  Need 'c() function here:
X1
```
